The Iris dataset consists of 150 records with 6 features: Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm, and Species. It is a balanced dataset with three species of iris flowersâ€”Iris-setosa, Iris-versicolor, and Iris-virginicaâ€”each represented by 50 samples. All the data types are appropriate, with numeric types for measurements and a categorical type for species. There are no missing values in the dataset, indicating clean and complete data. The statistical summary shows that petal length and width vary significantly across species, making them more useful for classification than sepal measurements. Sepal lengths range from 4.3 to 7.9 cm, while petal lengths range from 1.0 to 6.9 cm, showing greater variation. The dataset is widely used in machine learning for demonstrating classification algorithms due to its simplicity, balanced class distribution, and clearly separable feature patterns, especially in the petal dimensions.

```python
Y = df['Species']
X = df.drop(columns = ['Id','Species'])



```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.80)
```

```python
X_test.shape
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#Training
model.fit(X_train, Y_train)

# Testing
y_pred = model.predict(X_test)

#Compare
from sklearn.metrics import accuracy_score
score = accuracy_score(Y_test, y_pred)
accuracy = score*100
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix

#print the confusion matrix
print("Confusion Matrix")
cm = confusion_matrix(Y_test, y_pred)
print(cm)
```

```python
plt.figure(figsize = (6, 4))
sns.heatmap(cm,annot = True, fmt = 'd', cmap = 'Blues', cbar = False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.metrics import classification_report
# Generate classification report
report = classification_report(Y_test, y_pred)
print(report)
```
